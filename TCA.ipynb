```python
# import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns

# Create crosstab of counts for SeniorCitizen vs Churn
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert counts to % of total for each SeniorCitizen group
ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
plt.figure(figsize=(4,4))
ax = ct_percent.plot(kind='bar', stacked=True, color=['#66b3ff', '#ff9999'], width=0.6)

# Add % labels inside each stacked bar
for container in ax.containers:
    labels = [f'{v.get_height():.1f}%' if v.get_height() > 0 else '' for v in container]
    ax.bar_label(container, labels=labels, label_type='center', color='white', fontsize=9)

# Titles and labels
plt.title("Churn by SeniorCitizen (as % of Total)")
plt.ylabel("Percentage (%)")
plt.xlabel("SeniorCitizen")
plt.legend(title="Churn", bbox_to_anchor=(0.95,0.75),loc = "upper left")
sns.despine()
plt.tight_layout()
plt.show()
```

comperatively  a greater percentage of people in senior citizen catagory have churned

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of columns to plot
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Create subplots â€” adjust number of rows/columns as needed
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(12, 10))
axes = axes.flatten()  # Flatten 2D array of axes for easy iteration

# Loop through each column and create a countplot
for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, ax=axes[i], palette='Set2',hue = df["Churn"])
    axes[i].set_title(f'{col} Count', fontsize=11)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Remove empty subplots if number of columns < grid size
for j in range(len(cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

The plots show that customers lacking additional services such as OnlineSecurity, TechSupport, and DeviceProtection have a higher churn rate.
Fiber optic users tend to churn more than DSL users.
Customers with basic phone and internet services only are more likely to leave, while those with multiple or bundled services are more loyal.
Overall, service engagement is negatively correlated with churn â€” the more services a customer uses, the less likely they are to leave.
